```python
import seaborn as sns
import matplotlib.pyplot as plt
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
from IPython.display import FileLink
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,MinMaxScaler
from lightgbm import LGBMRegressor
from bs4 import BeautifulSoup
import requests
```

```python
def check_status_code(url):
    r=requests.get(url)
    print(r.status_code)
```

```python
def get_page_links(page_start,page_end):
    pages=np.arange(page_start,page_end)
    page_links=[]  
    for page in pages:
        page_link="https://www.emlakjet.com/kiralik-konut/istanbul/{page}/".format(page=page)
        page_links.append(page_link)
    return page_links
```

```python
page_links=get_page_links(1,51)
```

```python
def getHouseInfo(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.text,"html.parser")
    #print(soup.prettify())
    return soup 
```

```python

announs=[]
for i in range(len(page_links)):
    url=page_links[i]
    soup=getHouseInfo(url)
    
    for j in range(30):
        link="https://www.emlakjet.com"+soup.find_all("div",class_="styles_listingItem__1asTK")[j].find("a").get("href")
        announs.append(link)
```

```python
len(announs)
```

```python
url=announs[111]
r=requests.get(url)
soup=BeautifulSoup(r.content,"html.parser")
```

```python
nos=[]
ann_dates=[]
types=[]
categories=[]
const_types=[]
net_m2s=[]
brut_m2s=[]
rooms=[]
ages=[]
floors=[]
no_of_floors=[]
heatings=[]
dues=[]
deposits=[]
no_of_bathrooms=[]
no_of_wcs=[]
in_sites=[]
furnitures=[]
prices=[]

for i in announs: 
    url=i
    r=requests.get(url)
    soup=BeautifulSoup(r.content,"html.parser")
    
    try:
        
            no=soup.find_all(class_="styles_tableHalfRow__3zO2j")[0].find_all(class_="styles_tableColumn__2x6nG")[1].text
            nos.append(no)
            date=soup.find_all(class_="styles_tableHalfRow__3zO2j")[1].find_all(class_="styles_tableColumn__2x6nG")[1].text
            ann_dates.append(date)
            typ=soup.find_all(class_="styles_tableHalfRow__3zO2j")[3].find_all(class_="styles_tableColumn__2x6nG")[1].text
            types.append(typ)
            cat=soup.find_all(class_="styles_tableHalfRow__3zO2j")[4].find_all(class_="styles_tableColumn__2x6nG")[1].text
            categories.append(cat)
            const_type=soup.find_all(class_="styles_tableHalfRow__3zO2j")[5].find_all(class_="styles_tableColumn__2x6nG")[1].text
            const_types.append(const_type)
            net_m2=soup.find_all(class_="styles_tableHalfRow__3zO2j")[6].find_all(class_="styles_tableColumn__2x6nG")[1].text
            net_m2s.append(net_m2)
            brut_m2=soup.find_all(class_="styles_tableHalfRow__3zO2j")[7].find_all(class_="styles_tableColumn__2x6nG")[1].text
            brut_m2s.append(brut_m2)
            room=soup.find_all(class_="styles_tableHalfRow__3zO2j")[9].find_all(class_="styles_tableColumn__2x6nG")[1].text
            rooms.append(room)
            age=soup.find_all(class_="styles_tableHalfRow__3zO2j")[10].find_all(class_="styles_tableColumn__2x6nG")[1].text
            ages.append(age)
            floor=soup.find_all(class_="styles_tableHalfRow__3zO2j")[11].find_all(class_="styles_tableColumn__2x6nG")[1].text
            floors.append(floor)
            no_floor=soup.find_all(class_="styles_tableHalfRow__3zO2j")[12].find_all(class_="styles_tableColumn__2x6nG")[1].text
            no_of_floors.append(no_floor)
            heat=soup.find_all(class_="styles_tableHalfRow__3zO2j")[13].find_all(class_="styles_tableColumn__2x6nG")[1].text
            heatings.append(heat)
            due=soup.find_all(class_="styles_tableHalfRow__3zO2j")[14].find_all(class_="styles_tableColumn__2x6nG")[1].text
            dues.append(due)
            deposit=soup.find_all(class_="styles_tableHalfRow__3zO2j")[15].find_all(class_="styles_tableColumn__2x6nG")[1].text
            deposits.append(deposit)
            no_of_bathroom=soup.find_all(class_="styles_tableHalfRow__3zO2j")[16].find_all(class_="styles_tableColumn__2x6nG")[1].text
            no_of_bathrooms.append(no_of_bathroom)
            no_of_wc=soup.find_all(class_="styles_tableHalfRow__3zO2j")[17].find_all(class_="styles_tableColumn__2x6nG")[1].text
            no_of_wcs.append(no_of_wc)
            in_site=soup.find_all(class_="styles_tableHalfRow__3zO2j")[18].find_all(class_="styles_tableColumn__2x6nG")[1].text
            in_sites.append(in_site)
            furniture=soup.find_all(class_="styles_tableHalfRow__3zO2j")[19].find_all(class_="styles_tableColumn__2x6nG")[1].text
            furnitures.append(furniture)
            price=soup.find(class_="styles_price__1e65F").text
            prices.append(price)
            
            
            
    except (IndexError,TypeError,AttributeError):
            pass; 
```

```python
d={"price":prices,"ann_date":ann_dates,"type":types,"category":categories,"const_type":const_types,
  "net_m2":net_m2s,"brut_m2":brut_m2s,"room":rooms,"age":ages,"floor":floors,"no_of_floor":no_of_floors,
  "heating":heatings,"due":dues,"deposit":deposits,"no_of_bathroom":no_of_bathrooms,"no_of_wc":no_of_wcs,
  "in_site":in_sites,"furniture":furnitures,"price":prices}


data=pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in d.items() ]))
```
