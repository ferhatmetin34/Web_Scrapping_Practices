```python
from bs4 import BeautifulSoup
import requests
import pandas as pd
import numpy as np
pd.set_option('display.max_columns', None)
pd.set_option('display.max_rows',None)
from IPython.display import FileLink
```

```python
def check_status_code(url):
    r=requests.get(url)
    print(r.status_code)
```

```python
check_status_code("https://www.truecar.com/used-cars-for-sale/listings/")
```

```python
def get_page_links(page_start,page_end):
    pages=np.arange(page_start,page_end)
    page_links=[]
    
    for page in pages:
        page_link="https://www.truecar.com/used-cars-for-sale/listings/?page={page}&sort[]=best_match".format(page=page)
        page_links.append(page_link)
    return page_links
```

```python
def getCarInfo(url):
    r=requests.get(url)
    soup=BeautifulSoup(r.text,"html.parser")
    #print(soup.prettify())
    return soup 
```

```python
page_links=get_page_links(1,332)
announs=[]
for i in range(len(page_links)):
    url=page_links[i]
    soup=getCarInfo(url)
    
    for j in range(29):
        link="https://www.truecar.com"+soup.find_all(class_="linkable order-2 vehicle-card-overlay")[j].get("href")
        announs.append(link)
```

```python
len(announs)
```

```python
prices=[]
ext_colors=[]
int_colors=[] 
MPGs=[]
engines=[] 
transmissions=[] 
drive_types=[]
fuel_types=[]
mileages=[]
opt_levels=[] 
styles=[]  
locations=[]

for i in announs: 
    url=i
    r=requests.get(url)
    soup=BeautifulSoup(r.content,"html.parser")
    
    try:
           
            price=soup.find(class_="label-block-text").text
            prices.append(price)
            
            exterior_col=soup.find_all(class_="_1l9na3c align-self-center col")[0].find("li").text
            ext_colors.append(exterior_col)
            
            int_col=soup.find_all(class_="_1l9na3c align-self-center col")[2].find("li").text
            int_colors.append(int_col)
            
            mpg=soup.find_all(class_="_1l9na3c align-self-center col")[3].find("li").text
            MPGs.append(mpg)
            
            eng=soup.find_all(class_="_1l9na3c align-self-center col")[4].find("li").text
            engines.append(eng)
            
            transmission=soup.find_all(class_="_1l9na3c align-self-center col")[5].find("li").text
            transmissions.append(transmission)
            
            drive_type=soup.find_all(class_="_1l9na3c align-self-center col")[6].find("li").text
            drive_types.append(drive_type)
            
            fuel_type=soup.find_all(class_="_1l9na3c align-self-center col")[7].find("li").text
            fuel_types.append(fuel_type)
            
            mileage=soup.find_all(class_="_1l9na3c align-self-center col")[8].find("li").text
            mileages.append(mileage)
            
            opt_level=soup.find_all(class_="_1l9na3c align-self-center col")[9].find("li").text
            opt_levels.append(opt_level)
            
            style=soup.find_all(class_="_1l9na3c align-self-center col")[1].find("li").text
            styles.append(style)
        
            loc=soup.find(class_="_6c9ih0 _b9xhv wrappable-2 text-md margin-top-1").text
            locations.append(loc)
              

    except (IndexError,TypeError,AttributeError):
            pass; 
```

```python
d={"price":prices,"exterior_col":ext_colors,"interior_col":int_colors,"MPG":MPGs,
    "engine":engines,"transmission":transmissions,"drive_type":drive_types,"fuel_type":fuel_types,
    "mileage":mileages,"option_level":opt_levels,"style":styles,"location":locations}


data=pd.DataFrame(dict([ (k,pd.Series(v)) for k,v in d.items() ]))   
```
